```python
# convert .avi to .mp4 to allow embedding in jupyter
def convert_avi_to_mp4(avi_file_path, output_name):
    os.popen("ffmpeg -y -i '{input}' -ac 2 -b:v 2000k -c:a aac -c:v libx264 -b:a 160k -vprofile high -bf 0 -strict experimental -f mp4 '{output}.mp4'".format(input = avi_file_path, output = output_name))
    return True

convert_avi_to_mp4('/home/tomhagley/Documents/SLEAPProject/octagon_solo/analysis/test.avi', '/home/tomhagley/Documents/SLEAPProject/octagon_solo/analysis/test')
```

```python
def color_video_frame(video_path, stim_frames, trial, color_frame_delay):
    
    # save movie frame
    cap = cv2.VideoCapture(video_path)
    cap.set(cv2.CAP_PROP_POS_FRAMES, stim_frames.iloc[trial]  - color_frame_delay + 5)
    ret, image = cap.read()
    file_name = f"frame {stim_frames.iloc[trial]}.jpg"
    cv2.imwrite(file_name, image)
    print("frame saved")

    # display movie frame
    img = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)
    fig, ax = plt.subplots()
    ax.imshow(img, cmap='gray')
    plt.show()

    return file_name
```

```python
plt.rcParams["figure.figsize"] = (7,7)
color_video_frame(video_path=video_color_filename, stim_frames=colorstims, trial=trial, color_frame_delay=color_frame_delay)
```
